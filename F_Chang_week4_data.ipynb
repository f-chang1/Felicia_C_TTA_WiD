```python
import pandas as pd
import matplotlib.pyplot as plt

hol_data = pd.read_csv("holiday_data.csv")
hol_data.set_index("destination", inplace=True)
```

```python
# Task 1
print("Number of rows = " + str(hol_data.shape[0]))
print("Number of columns = " + str(hol_data.shape[1]))
```

```python
# Task 2
print(hol_data.iloc[2:8])
```

```python
# Task 3
print("The mean number of all-inclusive hotels is " + str(hol_data.all_inclusive.mean()))
```

```python
# Task 4
print("The lowest scoring destination is " + str(hol_data.feedback_score.idxmin()))
```

```python
# Task 5
print("The highest scoring destination is " + str(hol_data.feedback_score.idxmax()))
```

```python
# Task 6
print("The following destionations have more than 9 all-inclusive hotels:")
print(hol_data[hol_data.all_inclusive > 9].index.tolist())
```

```python
# Task 7
print("Destinations with a feedback score above 8:")
print(hol_data[hol_data.feedback_score > 8].feedback_score)
```

```python
# Task 8
print("Destinations with a feedback score below 2:")
print(hol_data[hol_data.feedback_score < 2].feedback_score)
```

```python
hol_data.plot.scatter(x="all_inclusive", y="feedback_score")
ax1 = plt.xlabel("Number of all-inclusive hotels")
plt.ylabel("Feedback score (/10)")
print('The Pearson correlation coefficient between the number of all-inclusive hotels and feedback score =',
      hol_data.corr().loc['all_inclusive', 'feedback_score'])
print('There is therefore no evidence of correlation between the number of all-inclusive hotels and feedback score.')
```

```python
plt.figure
ax2 = hol_data["feedback_score"].sort_values().plot(kind='barh')
plt.xlabel("Feedback score (/10)")
plt.ylabel("Destination")
plt.title("Destinations by feedback score")
print("Visualising destinations by highest feedback score:")
```
