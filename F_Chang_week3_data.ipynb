```python
# Python task 1

def number_ent():
    in_num = input("Please enter a number.")
    valid = False
    while not valid:
        try:
            num = float(in_num)
            valid = True
        except ValueError:
            in_num = input("Please make sure your input is a number. Try again?")
    return num

my_file = open("Numbers.txt", "w")

for i in range(4):
    num = number_ent()
    my_file.write(str(num))
    my_file.write("\n")

my_file.close()

with open ("Numbers.txt", "r") as myfile:
    data = myfile.read().replace('\n', ',')

myfile.close()

print(data)

my_file = open("Numbers.txt", "a")

my_file.write("test append")

my_file.close()

my_file = open("Numbers.txt", "r")
data = my_file.read()

myfile.close()

print(data)
```

```python
#Python task 2 and extension

grade_boundaries = {"A*": 90, "A":80, "B":70, "C":60, "F":0}

def percent_ent():
    # asks for a user input percentage and checks that it is a number
    in_num = input("Please enter your percentage mark.")
    valid = False
    while not valid:
        try:
            num = float(in_num)
            valid = True
        except ValueError:
            in_num = input("Please make sure your input is a number. Try again?")
    return num

def mark_grade(mark, boundaries):
    # takes a mark (float) and set of grade boundaries (dict) and returns a letter grade
    grade_list = sorted(boundaries, key=boundaries.get, reverse=True)
    for grade in grade_list:
        if mark >= boundaries[grade]:
            return grade
    return "F"

def test_grade(actual, boundaries):
    # takes a grade (str) and set of grade boundaries (dict), asks for a target grade and compares the achieved grade to the target grade
    grade_list = sorted(boundaries, key=boundaries.get, reverse=True)
    valid = False
    grade = input("Please enter your target grade from the following list: " + str(grade_list))
    while not valid:
        if grade in grade_list:
            valid = True
        else:
            grade = input("Please try again, making sure your target grade from the following list: " + str(grade_list))
    if grade_list.index(actual) < grade_list.index(grade):
        print("Congratulations, you exceeded your target grade.")
    elif grade_list.index(actual) == grade_list.index(grade):
        print("Congratulations, you achieved your target grade.")
    else:
        print("Unfortunately you did not achieve your target grade.")

percent = percent_ent()
grade = mark_grade(percent, grade_boundaries)

print("Your grade is " + grade)

test_grade(grade, grade_boundaries)
```

```python
# Numpy extension tasks

import numpy as np

# Extension task 1
base = np.arange(1,4)
a = np.repeat(base, 3)
b = np.tile(base, 3)
ext_1 = np.concatenate((a,b))
print(ext_1)

# For extension tasks 2 and 3
a = np.arange(1,6)
b = np.arange(4,10)

# Extension task 2
ext_2 = b[~np.in1d(b, a)]
print(ext_2)

# Extension task 3
ext_3 = np.concatenate((a[(a>=5) & (a<=10)], b[(b>=5) & (b<=10)])).sum()
print(ext_3)
```
